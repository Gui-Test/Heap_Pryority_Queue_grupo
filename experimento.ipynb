```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
dados =pd.read_csv("C:/Users/rened/OneDrive/Documentos/códigos/trabalho_heap/Heap_Pryority_Queue_grupo/dados.csv")
dados
```

```python
# Perform linear fit
x =  dados.NumeroInserido
y = dados.Iterações
coefficients = np.polyfit(x, y, 2)
print("Linear Fit Coefficients:", coefficients)

# Create polynomial function
p = np.poly1d(coefficients)

plt.scatter(x, y, label='Data Points')
plt.plot(x, p(x), label='Linear Fit', color='red')
plt.legend()
plt.show()
```
